```python
def aa_perc(protein, aminoacid=['A', 'I', 'L', 'M', 'F', 'W', 'Y', 'V'], sig_figs=1):
    protein=protein.upper()
    length= len(protein)
    aacount=0
    for aa in aminoacid:
        aacount= aacount+protein.count(aa)
    percent= aacount/length*100
    return round(percent, sig_figs)
    
```
